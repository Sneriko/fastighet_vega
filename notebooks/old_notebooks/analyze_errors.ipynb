```python
import pandas as pd

df = pd.read_csv('/home/erik/Riksarkivet/Projects/fastighet/output/error_list_2_it.txt', delimiter='\t', names=['batch', 'batch_pnr', 'gt', 'pred', 'sc_pred'])
```

```python
batches = df.groupby('batch')
list_of_batches = df.batch.unique()
no_of_batches = len(list_of_batches)

errors_by_batch = []

for batch in list_of_batches:
    df_batch = batches.get_group(batch)
    errors_by_batch.append((str(batch), len(df_batch)))

errors_by_batch = sorted(errors_by_batch, key=lambda x: x[1])
errors_by_batch.reverse()

print(6227/no_of_batches)

with open('/home/erik/Riksarkivet/Projects/fastighet/output/errors_by_batch', 'w') as f:
    for b in errors_by_batch:
        f.write(b[0] + '\t' + str(b[1]) + '\n')
    
```
