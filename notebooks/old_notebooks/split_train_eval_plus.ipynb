```python
import os
from glob import glob
from pathlib import Path
import pandas as pd
from itertools import groupby

dataset_path = '/home/erik/Riksarkivet/Projects/fastighet/data/htr_test_set_1_million'

batches_eval = glob(os.path.join(dataset_path, '*/'))
batches_eval.sort()
batches_eval = [Path(x).name for x in batches_eval]
```

```python
key_func = lambda x: x[9:17]
```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_1_million_plus/gt_all_plus.txt', 'r') as f, open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_1_million_plus/gt_all_eval_plus.txt', 'w') as f_eval, open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_1_million_plus/gt_all_train_plus.txt', 'w') as f_train:
    gts = f.readlines()

    batches_all = groupby(gts, key_func)

    batches_all = [list(b[1]) for b in batches_all]

    for batch in batches_all:
        for inst in batch:
            if inst[0:8] in batches_eval:
                f_eval.write(inst)
            else:
                f_train.write(inst)
```
