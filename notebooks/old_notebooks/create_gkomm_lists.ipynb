```python
import pandas as pd
import json
```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/output/pipeline_htr_gkommun.json', 'r') as f:
    htr_dict = json.load(f)
```

```python


gkomm_dict = {}

for batch in htr_dict.keys():
    gkomms = []
    
    for page in htr_dict[batch].keys():

        if htr_dict[batch][page] is None:
            continue

        for inst in htr_dict[batch][page]:
            for gt in inst['gts']
                gkomm = gt.split(';')[0]
                if gkomm not in gkomms:
                    gkomms.append(gkomm)

    gkomm_dict[batch] = gkomms
```
