```python
import random
```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_40_batches/gt.txt', 'r') as f, open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_40_batches/eval.txt', 'w') as f_eval, open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_40_batches/train.txt', 'w') as f_train:
    gts = f.readlines()
    random.shuffle(gts)

    cut = int(0.1 * len(gts))

    eval = gts[0:cut]
    train = gts[cut:]

    for inst in eval:
        f_eval.write(inst)

    for inst in train:
        f_train.write(inst)

```
