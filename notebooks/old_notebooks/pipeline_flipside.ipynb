```python
import json
import copy
```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/output/iteration3_sc_epoch3.json', 'r') as f:
    eval_dict = json.load(f)
```

```python
for batch in eval_dict.keys():

    for i, page in enumerate(eval_dict[batch].keys()):

        if eval_dict[batch][page] is not None and eval_dict[batch][page][0]['det_prob'] is None:
            
            page_number = int(page.split('_')[1])
            previous_page = batch + '_' + (str(page_number - 1)).zfill(8)
            next_page = batch + '_' + (str(page_number + 1)).zfill(8)
            
            if page_number % 2 == 0:
                eval_dict[batch][page] = copy.deepcopy(eval_dict[batch][previous_page])
                for inst in eval_dict[batch][page]:
                    inst['cop_from_flipside'] = True
            else:
                print(page)
                eval_dict[batch][page] = copy.deepcopy(eval_dict[batch][next_page])
                for inst in eval_dict[batch][page]:
                    inst['cop_from_flipside'] = True

        else:
            continue

```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/output/iteration3_flipside_epoch3.json', 'w', encoding='utf8') as f:
    j = json.dumps(eval_dict, indent = 4, ensure_ascii=False)
    f.write(str(j))
```
