```python
import pandas as pd
from pathlib import Path
from glob import glob
import os
import json
```

```python
df_million_single = pd.read_csv('/home/erik/Riksarkivet/Projects/fastighet/data/miljonsetet/miljon_testet_facit_enkel_FNR.txt', delimiter='\t', index_col=2, encoding='latin-1', dtype=str)
df_million_multiple = pd.read_csv('/home/erik/Riksarkivet/Projects/fastighet/data/miljonsetet/miljon_testet_facit_flera_FNR_datum_stigande.txt', index_col=2, delimiter='\t', encoding='latin-1', dtype=str)
df_million_conc = pd.concat([df_million_single, df_million_multiple])

#drop duplicates, add here
```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/output/iteration3_htr_epoch3.json', 'r') as f:
    dict_htr = json.load(f) 
```

```python
#Skriv om så att det blir rätt här, rätt socken för varje gt

for i, batch in enumerate(dict_htr.keys()):

    for page in dict_htr[batch].keys():

        try:
            rows = df_million_conc.loc[page]
        except:
            print(page + ' not_indexed')
            continue

        gts_new =  []

        if isinstance(rows, pd.Series):
            socken = rows['GKOMMUN']
            trakt = rows['GTRAKT']
            block = rows['GBLOCK']
            enhet = rows['GENHET']
            fnr = rows['LMVID']
            gt_string = socken + ';' + trakt + ';' + block + ';' + enhet
            gts_new.append((gt_string, str(fnr)))
        else:
            for index, row in rows.iterrows():
                socken = row['GKOMMUN']
                trakt = row['GTRAKT']
                block = row['GBLOCK']
                enhet = row['GENHET']
                fnr = row['LMVID']
                gt_string = socken + ';' + trakt + ';' + block + ';' + enhet
                gts_new.append((gt_string, str(fnr)))

        

        try:
            for inst in dict_htr[batch][page]:
                inst['gts'] = gts_new
        except:
            print(page)

    print(i)
```

```python
batches = glob(os.path.join('/home/erik/Riksarkivet/Projects/fastighet/data/htr_test_set_1_million', '**'))

for i, batch in enumerate(batches):

    batch_name = Path(batch).name

    imgs = glob(os.path.join(batch, '**'))

    for img in imgs:
        img_name = '_'.join(Path(img).name.split('_')[0:2])

        try:
            rows = df_million_conc.loc[img_name]
        except:
            pass

        gts_new =  []

        if isinstance(rows, pd.Series):
            socken = rows['GKOMMUN']
            trakt = rows['GTRAKT']
            block = rows['GBLOCK']
            enhet = rows['GENHET']
            fnr = rows['LMVID']
            gt_string = socken + ';' + trakt + ';' + block + ';' + enhet
            gts_new.append((gt_string, str(fnr)))
        else:
            for index, row in rows.iterrows():
                socken = row['GKOMMUN']
                trakt = row['GTRAKT']
                block = row['GBLOCK']
                enhet = row['GENHET']
                fnr = row['LMVID']
                gt_string = socken + ';' + trakt + ';' + block + ';' + enhet
                gts_new.append((gt_string, str(fnr)))

        

        try:
            for inst in dict_htr[batch_name][img_name]:
                inst['gts'] = gts_new
        except:
            print(img_name)

    print(i)

            

        

```
