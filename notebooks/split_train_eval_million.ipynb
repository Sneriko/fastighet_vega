```python
import random
from itertools import groupby
```

```python
key_func = lambda x: x[9:17]
```

```python
with open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_1_million/gt_all.txt', 'r') as f, open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_1_million/gt_all_eval.txt', 'w') as f_eval, open('/home/erik/Riksarkivet/Projects/fastighet/data/htr_training_set_1_million/gt_all_train.txt', 'w') as f_train:
    gts = f.readlines()

    batches = groupby(gts, key_func)

    batches = [list(b[1]) for b in batches]

    random.shuffle(batches)

    cut = int(0.1 * len(batches))

    eval_batches = batches[0:cut]
    train_batches = batches[cut:]
    
    for batch in eval_batches:
        for inst in batch:
            f_eval.write(inst)

    for batch in train_batches:
        for inst in batch:
            f_train.write(inst)
    
```
