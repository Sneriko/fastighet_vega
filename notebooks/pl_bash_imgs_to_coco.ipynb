```python
from glob import glob
import mmcv
import os
from pathlib import PurePath
import json
```

```python
#ta två batcher och testa dig igenom på dom

imgs = glob('/ceph/hpc/home/euerikl/projects/fastighet/data/all_batches/**/*.tif')
imgs.sort()
imgs = imgs[0:100]
```

```python
coco = dict()
coco['images'] = list()



for i, img_p in enumerate(imgs):
    
    img_dict = dict()
    
    img = mmcv.imread(img_p)
    
    height = img.shape[0]
    width = img.shape[1]
    file_name = ('/').join(PurePath(img_p).parts[-2:])

    img_dict['width'] = width
    img_dict['height'] = height
    img_dict['id'] = i
    img_dict['file_name'] = file_name

    coco['images'].append(img_dict)
    
```

```python
with open('/ceph/hpc/home/euerikl/projects/fastighet/output/coco_ex.json', 'w', encoding='utf8') as f:
    coco_str = json.dumps(coco, indent = 4, ensure_ascii=False)
    f.write(str(coco_str))
```
