```python
import os
import mmcv
import random
```

```python
path_to_imgs = '/ceph/hpc/scratch/user/euerikl/data/lagfart/cropped_imgs_ind_batches/cropped_imgs_ind_batches'
```

```python
with open('/ceph/hpc/home/euerikl/projects/fastighet/data/lagfart/fn_iter2/gt_all_2.txt', 'r') as f:
    
    lines = f.readlines()
    valid_lines = list()
    invalid_lines = list()
    
    for i, line in enumerate(lines):
        img_p = line.split('|')[0]
        img_p = os.path.join(path_to_imgs, img_p)
        
        if os.path.exists(img_p):
            valid_lines.append(line)
        else:
            invalid_lines.append(line)

        if i % 1000 == 0:
            print(i)
        
```

```python
random.shuffle(valid_lines)

cutoff = int(0.95 * len(valid_lines))
train_iter2 = valid_lines[0:cutoff]
eval_iter2 = valid_lines[cutoff:]

with open('/ceph/hpc/home/euerikl/projects/fastighet/data/lagfart/fn_iter2/train.txt', 'w') as f_t:
    for l in train_iter2:
        f_t.write(l)

with open('/ceph/hpc/home/euerikl/projects/fastighet/data/lagfart/fn_iter2/eval.txt', 'w') as f_e:
    for l in eval_iter2:
        f_e.write(l)

```
