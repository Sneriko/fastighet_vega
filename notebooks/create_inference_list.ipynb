```python
import json
from pathlib import Path
import os
```

```python
with open('/ceph/hpc/home/euerikl/projects/fastighet/data/prop_det_40b_train/result.json', 'r') as f:   
    coco = json.load(f)
```

```python
#"file_name": "images/14/35aecd38-10014756_00000411.jpg"

for img_dict in coco['images']:
    old_path = img_dict['file_name']
    img_name = Path(old_path).stem.split('-')[1] + '.tif'
    batch = img_name.split('_')[0]
    new_path = os.path.join(batch, img_name)
    img_dict['file_name'] = new_path
```

```python
with open('/ceph/hpc/home/euerikl/projects/fastighet/data/prop_det_40b_train/test_dist.json', 'w', encoding='utf8') as f:
    s = json.dumps(coco, indent = 4, ensure_ascii=False)
    f.write(str(s))
```
