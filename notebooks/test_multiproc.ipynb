```python
import multiprocessing as mp
from glob import glob
import mmcv
from pathlib import PurePath
```

```python
imgs = glob('/ceph/hpc/home/euerikl/projects/fastighet/data/pipeline_test_data/batches/**/*.tif')
imgs.sort()

coco = dict()

coco['images'] = list()
```

```python
def create_inst(img_p):
    img_dict = dict()
    
    img = mmcv.imread(img_p)
    
    height = img.shape[0]
    width = img.shape[1]
    file_name = ('/').join(PurePath(img_p).parts[-2:])

    img_dict['width'] = width
    img_dict['height'] = height
    #img_dict['id'] = i
    img_dict['file_name'] = file_name

    return img_dict
```

```python
mp.freeze_support()
p = mp.Pool()
scores = p.map(create_inst, imgs)
p.close()
p.join()
```

```python
scores = sorted(scores, key=lambda x: x['file_name'])
for i, inst in enumerate(scores):
    inst['id'] = i
```
